```python
# From the above result, we can  see that "Hardtop" has the Highest_Average Horse_Power but there only 8 in Numbers.
```

***CONCATENATE(TABLE_JOIN) IN PYTHON***

```python
# Using Table_Join to "Extact" Certain Information from the Original table to be use for quick reference and analysis
# Checking for Cars that were Made by "Audi" whoes Fuel_type is "Gas" and using "Table_JOin" to "Save the Output"
# For future Reference and Further Analysis as the Focus may be.
# And using "Table_JOin. We have to Assign a "Variable" to any "Output" to make the workload easy
```

```python
genius[genius["fuel_type"]=="diesel"]
```

```python
# Assign a Variable
```

```python
s = genius[genius["fuel_type"]=="diesel"]
```

```python
s
```

```python
# Next_step is to find the cars Make by "Audi"
```

```python
genius[genius["make"]=="audi"]
```

```python
# Assign a Variable to our Second Output
```

```python
t = genius[genius["make"]=="audi"]
```

```python
t
```

```python
# Now let's join the Two newly created Tables together using "Concatenate" and Pass them through a "LIST" Datastructure
```

```python
pd.concat([s,t])
```

```python
# Reseting the Index
```

```python
pd.concat([s,t]).reset_index()
```

```python
# Drop the "Original Index"
```

```python
pd.concat([s,t]).reset_index(drop = True)
```
